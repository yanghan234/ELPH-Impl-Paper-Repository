```python
import json
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
# from sklearn.linear_model import LinearRegression
mpl.rcParams['figure.dpi']= 200
%config InlineBackend.figure_format = 'svg'
# plt.rcParams['font.size'] = 15
# mpl.rcParams['axes.linewidth'] = 2

RYTOEV = 13.605698 # Rydberg to eV
ryd2mev = RYTOEV*1000
```

```python
# Kresse 
refTs1 = [0,100,200,300,400,500,600,700]
refZPRs1 = [ 0.0,
           -0.25642263859377934,
           -1.3704848367780542,
           -5.5703109909213815,
           -16.970253042302488,
           -35.571759706393664,
           -60.34334556693065,
           -89.22928336874638]

# Antonius 2014
refTs2 = range(0,651,50)
refZPRs2 = np.array([7.327569194997988,
                    7.327670546852875,
                    7.326959593400206,
                    7.325439315576886,
                    7.323109713382916,
                    7.319156991042284,
                    7.314393453862549,
                    7.308819101843709,
                    7.300002980936909,
                    7.290373064254095,
                    7.279124498829982,
                    7.266251322790753,
                    7.251759498010225,
                    7.2364523869852295])
refZPRs2 = (refZPRs2 - refZPRs2[0])*1000
```

```python
def read_ZPR(prefix,skipline=4):
    fmfname = prefix +"/reFMselfen_DFT_adi.tab"
    vbm_fm = 0
    cbm_fm = 0
    with open(fmfname,"r") as f:
        for i in range(skipline):
            f.readline()
        line = f.readline()
        vbm_fm = float(line.split()[3])
        line = f.readline()
        cbm_fm = float(line.split()[3])
    print(fmfname)
    print([vbm_fm,cbm_fm])
    dwfname = prefix + "/reDWselfen_DFT.tab"
    vbm_dw = 0
    cbm_dw = 0
    with open(dwfname,"r") as f:
        for i in range(skipline):
            f.readline()
        line = f.readline()
        vbm_dw = float(line.split()[3])
        line = f.readline()
        cbm_dw = float(line.split()[3])
    
    vbm_zpr = vbm_fm + vbm_dw
    cbm_zpr = cbm_fm + cbm_dw
    gap_zpr = cbm_zpr - vbm_zpr
    return [vbm_fm,vbm_dw,vbm_zpr,cbm_fm,cbm_dw,cbm_zpr,gap_zpr]
```

```python
Ts = [0,100,200,300,400,500,600,700]
GapZPRs = []
for t in Ts:
    prefix = "./60Ry/c.welph.save.pdep200.T%d"%(t)
    zpr = read_ZPR(prefix)
#     print(zpr)
    GapZPRs.append(zpr[-1])

GapZPRs = np.array(GapZPRs)
GapZPRs = GapZPRs - GapZPRs[0]
print(GapZPRs)
```

```python
fig = plt.figure(figsize=(8,4.5))

plt.plot(refTs1,refZPRs1,"s--",label="Karsai et al.")
plt.plot(refTs2,refZPRs2,"s--",label="Antonius et al.")
plt.plot(Ts,GapZPRs,'o-',label="This work")
plt.legend(fontsize=13,loc="lower left")
plt.xlabel("Temperature [Kelvin]",fontsize=13)
plt.ylabel("Direct gap renormalization [meV]",fontsize=13)
plt.show()
```
